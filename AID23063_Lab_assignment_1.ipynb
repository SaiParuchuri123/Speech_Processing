## Importing Modules

```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
import simpleaudio as sa
import librosa
```

## Importing and Reading a Speech File

```python

# Replace 'your_audio_file.wav' with the actual path to your audio file
audio_file = '/content/LJ001-0001.wav'
y, sr = librosa.load(audio_file);
```

## Basic Information Of Audio File

```python
print("Sampling rate (sr):", sr)
print("Number of Samples:", y.shape)
print("Audio duration (seconds):", librosa.get_duration(y=y, sr=sr))
```

## Slicing and Normalizing for first two seconds

```python

duration = 2
slice_samples = int(duration * sr)
sliced_signal = y[:slice_samples]
normalized_signal = librosa.util.normalize(sliced_signal)

plt.figure(figsize=(12, 6))



plt.subplot(2, 1, 1)
plt.plot(sliced_signal)
plt.title('Sliced Audio Signal (First 2 Seconds)')
plt.xlabel('Sample')
plt.ylabel('Amplitude')


plt.subplot(2, 1, 2)
plt.plot(normalized_signal)
plt.title('Normalized Audio Signal (First 2 Seconds)')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

# Amplification , De-amplification , Upsampling and Downsampling the audio signal

```python
# Amplification
amplified_signal = y * 2

# De-amplification
deamplified_signal = y * 0.5

# Upsampling
upsampled_signal = librosa.resample(y, orig_sr=sr, target_sr=sr * 2)

# Downsampling
downsampled_signal = librosa.resample(y, orig_sr=sr, target_sr=sr // 2)


plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.plot(y)
plt.title('Original Audio Signal')

plt.subplot(4, 1, 2)
plt.plot(amplified_signal)
plt.title('Amplified Audio Signal')

plt.subplot(4, 1, 3)
plt.plot(deamplified_signal)
plt.title('De-amplified Audio Signal')

plt.subplot(4, 1, 4)
plt.plot(upsampled_signal)
plt.title('Upsampled Audio Signal')


plt.tight_layout()
plt.show()

plt.figure(figsize=(7, 5))
plt.plot(downsampled_signal)
plt.title('Downsampled Audio Signal')
plt.tight_layout()
plt.show()
```
